```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X,y = make_blobs(n_samples=500,centers=3,cluster_std=0.60,random_state=0)

#visualize the data 
plt.scatter(X[:,0],X[:,-1],s=30)
plt.title("Generated data")
plt.show()
```

```python
#initailizing kmeans with k

kmeans = KMeans(n_clusters=3)

#fit the model

kmeans.fit(X)


#predict the cluster

y_kmeans = kmeans.predict(X)

#visualize the cluster 

plt.scatter(X[:,0],X[:,1],c=y_kmeans, s=30,cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.7,marker='X')
plt.title('KMeans')
plt.show()
```

```python
#elbow method - compute wcss(within cluster su=om of square)

wcss_list = []
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)


plt.plot(range(1,11),wcss_list,marker='o')
plt.xlabel("k")
plt.ylabel("WCSS")
plt.title("k vs MCSS")
plt.show()
```

```python
from sklearn.datasets import make_circles
X,y = make_circles(n_samples=500,factor=0.5,noise = 0.1,random_state=42)
plt.scatter(X[:,0],X[:,-1],s=30)
plt.title("Generated data")
plt.show()
```

```python
#initailizing kmeans with k

kmeans = KMeans(n_clusters=2)

#fit the model

kmeans.fit(X)


#predict the cluster

y_kmeans = kmeans.predict(X)

#visualize the cluster 

plt.scatter(X[:,0],X[:,1],c=y_kmeans, s=30,cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.7,marker='X')
plt.title('KMeans')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles


X,y = make_circles(n_samples=500,factor=0.5,noise = 0.05,random_state=42)
plt.scatter(X[:,0],X[:,-1],s=30)
plt.title("Generated data")
plt.show()



```

```python
dbscan = DBSCAN(eps=0.175,min_samples=10)
y_dbscan = dbscan.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=y_dbscan,s=50,cmap='viridis')
plt.title("DBSCAN")
plt.show()

```

```python
from sklearn.neighbors import NearestNeighbors

#compute the k -distance plot

neighbour=NearestNeighbors(n_neighbors=10)
neighbour_fit = neighbour.fit(X)
distance,indices =neighbour_fit.kneighbors(X)

#sort and plot the distances


distances =np.sort(distance[:,9],axis=0)
plt.plot(distances)
plt.xlabel("points")
plt.ylabel("Eplision")
plt.title("K-distance plot")
plt.show()
```
